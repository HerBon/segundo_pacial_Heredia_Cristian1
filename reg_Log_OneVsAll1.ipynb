```python
def sigmoid(z):  # Calcula la sigmoide de z
    return 1.0 / (1.0 + np.exp(-z))
```

```python
def lrCostFunction(theta, X, y, lambda_):
    m = y.size              # Inicializa algunos valores utiles
    if y.dtype == bool:     # convierte las etiquetas a valores enteros si son boleanos
        y = y.astype(int)   
    J = 0
    grad = np.zeros(theta.shape)    
    h = sigmoid(X.dot(theta.T))   
    temp = theta
    temp[0] = 0   
    J = (1 / m) * np.sum(-y.dot(np.log(h)) - (1 - y).dot(np.log(1 - h))) + (lambda_ / (2 * m)) * np.sum(np.square(temp))   
    grad = (1 / m) * (h - y).dot(X) 
    grad = grad + (lambda_ / m) * temp
    return J, grad
```

```python
def oneVsAll(X, y, num_labels, lambda_):   
    m, n = X.shape # filas y columas de "X"    
    all_theta = np.zeros((num_labels, n + 1))  # Matriz de Ceros para las Thetas, de m(filas) x (n+1)(columas)
    X = np.concatenate([np.ones((m, 1)), X], axis=1) # Agrega unos a la matriz X

    for c in np.arange(num_labels):
        initial_theta = np.zeros(n + 1)
        options = {'maxiter': 10000}
        res = optimize.minimize(lrCostFunction, 
                                initial_theta, 
                                (X, (y == (c+1)), lambda_), 
                                jac=True, 
                                method='BFGS',           #el mejor metodo BFGS, segundo TNC, 
                                options=options)       
        all_theta[c] = res.x
    return all_theta
```

```python
lambda_ = 0.1
all_theta = oneVsAll(X, y, num_labels, lambda_)
```

```python
def predictOneVsAll(all_theta, X): 
    m = X.shape[0];
    num_labels = all_theta.shape[0]
    p = np.zeros(m)

    X = np.concatenate([np.ones((m, 1)), X], axis=1)
    p = np.argmax(sigmoid(X.dot(all_theta.T)), axis = 1)

    return p+1
```
